# SCRAPING

```python
import requests


from requests import get

# Lien de la page à scraper
url = "https://unmade.breezy.hr/"

response = get(url)
source = None # Le code source de la page 

if response.status_code == 200 :
    # Si la requete s'est bien passee
    source = response.text
    
print(source)
```

# PARSING

```python
import bs4

from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(source, 'html.parser')

print(soup.prettify())
```

```python
database = [['job_name', 'country', 'city', 'offer_link']]

positions = soup.find_all('li', class_ = 'position transition')

for position in positions:
    
    #JOB_NAME
    job_name = position.find("h2").text
    #print(job_name)        
    
    # COUNTRY / CITY
    citycountry = position.find("span").text
    #print(citycountry)
    
    country = citycountry.split(",")[1]
    city = citycountry.split(",")[0]
    #print(city)
    
    # OFFER_LINK
    link = position.a.get('href')    
    offer_link = "https://unmade.breezy.hr/" + link
    #print(offer_link)

#    print(position.prettify())

    database.append([job_name, country, city, offer_link])
```
