# get_country
get_country returns the country of a city or the country itself if it is a country 
Whatever the input language, answer is returned in english

```python
from geopy.geocoders import Nominatim
 
geolocator = Nominatim(user_agent = "geoapiExercises")

def get_country(place):
    res = str(geolocator.geocode(place, language = "en"))
    country = res.split(',')[-1]
    return country
```

```python
soup = BeautifulSoup(source, 'html.parser')

print(soup.prettify())
```

```python
database = [['job_name', 'country', 'city', 'offer_link']]

offers = soup.find('div', class_ = 'block cms-block cms-block-jobs_by_department')
#print(offers)
positions = offers.find_all('li', class_ = 'jobs-list-item')

for position in positions:
    #print("______________")
    #print(position)
    
    #JOB_NAME
    job_name = position.find("h3").text.strip(' \n\t')
    #print(job_name)        
    
    # COUNTRY / CITY
    city = position.find("li", class_ = 'jobs-list-item-office').text.strip(' \n\t')
    country = get_country(city)

#     print(country)
#     print('the city is: '+ city)
    
    # OFFER_LINK
    link = position.a.get('href')    
    offer_link = "https://upmem.welcomekit.co" + link
    #print(offer_link)

    database.append([job_name, country, city, offer_link])
```
