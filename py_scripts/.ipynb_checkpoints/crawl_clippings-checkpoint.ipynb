# get_country
get_country returns the country of a city or the country itself if it is a country 
Whatever the input language, answer is returned in english

```python
from geopy.geocoders import Nominatim
 
geolocator = Nominatim(user_agent = "geoapiExercises")

def get_country(place):
    res = str(geolocator.geocode(place, language = "en"))
    country = res.split(',')[-1]
    return country
```

```python
import json
```

```python
database = [['job_name', 'country', 'city', 'offer_link']]

offers = soup.find('script', id = 'positionData').string#, class_ = 'list-container list')

positions = json.loads(offers) 


for position in positions:
    print("______________")
    print(position)
    
    #JOB_NAME
    job_name = position['jobOpeningName']
    #print(job_name)        
    
    # COUNTRY / CITY
    country = position['location']['country']
    city = position['location']['city']
    #print(country)
    #print('the city is: '+ city)
    
    # OFFER_LINK
    offer_link = "https://clippings.bamboohr.com/jobs/view.php?id=" + str(position['id'])
    #print(offer_link)

    database.append([job_name, country, city, offer_link])
```
