# get_country
get_country returns the country of a city or the country itself if it is a country 
Whatever the input language, answer is returned in english

```python
from geopy.geocoders import Nominatim
 
geolocator = Nominatim(user_agent = "geoapiExercises")

def get_country(place):
    res = str(geolocator.geocode(place, language = "en"))
    country = res.split(',')[-1]
    return country
```

```python
soup = BeautifulSoup(source, 'html.parser')

print(soup.prettify())
```

```python
database = [['job_name', 'country', 'city', 'offer_link']]

offers = soup.find('div', class_ = 'positions__table')

positions = offers.find_all('li')
#print(positions)

for position in positions:
    #print(position)
    
    #JOB_NAME
    job_name = position.find("p").text.strip(' \n\t')
    print(job_name)        
    
    # COUNTRY / CITY
    citycountry = position.find("p").next_sibling.next_sibling.next_sibling.next_sibling.text
    #print(citycountry)
    citycountry_list = citycountry.split(",")
 
    # Best case: the 2 fields are provided:
    if len(citycountry_list) == 2:
        country = get_country(citycountry_list[1]).strip(' \n\t')
        city = citycountry_list[0].strip(' \n\t')
    
    # But sometimes only the city or the country is provided: in that case, the length of citycountry is 1
    elif len(citycountry_list) == 1: 
        country = get_country(citycountry_list[0]).strip(' \n\t')
        city = ' '
        
    else:
        country = ' '
        city = ' '
    
    #print(country)
    #print('the city is: '+ city)
    
    # OFFER_LINK
    link = position.a.get('href')    
    offer_link = "https://drivenets.com" + link
    #print(offer_link)

#    print(position.prettify())

    database.append([job_name, country, city, offer_link])
```
