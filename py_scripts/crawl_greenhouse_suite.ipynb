# get_country
get_country returns the country of a city or the country itself if it is a country 
Whatever the input language, answer is returned in english

```python
from geopy.geocoders import Nominatim
 
geolocator = Nominatim(user_agent = "geoapiExercises")

def get_country(place):
    # Otherwise the function returns "United States"
    if place == "Asia Pacifica": 
        return "Asia"
    
    else:
        res = str(geolocator.geocode(place, language = "en"))
        country = res.split(',')[-1]
        return country
```

```python
def greenhouse_fetch_data(job):
    
        print(job['title'])

        #JOB NAME
        job_name = job['title']
        #print(job_name)
            
        #COUNTRY / CITY
        location = job['location']['name'].split(',')
        print(location)
        
        # Best case: the 2 fields are provided (or even 3):
        if len(location) >= 2:
            country = get_country(location[-1]).strip(' \n\t')
            city = location[0].strip(' \n\t')
        
        # But sometimes only the city or the country is provided: in that case, the length of citycountry is 1
        elif len(location) == 1: 
            country = get_country(location[0]).strip(' \n\t')
            city = ' '
        
        else:
            country = ' '
            city = ' '
            
        print(country)
        print(city)
    
        # OFFER_LINK
        offer_link = job['absolute_url']
        #print(offer_link)
        
        return([job_name, country, city, offer_link])
```

# MAIN

```python
import json

positions_int = json.loads(source)

positions = positions_int['departments']

print(positions)
```

```python
database = [['job_name', 'country', 'city', 'offer_link']]


for position in positions:
    
    # some categories can have "nested" tables of jobs in 'children' so we have to make sure we are not missing the jobs in children
    if position['children'] == []:
        for job in position['jobs']: 
            database.append(greenhouse_fetch_data(job))
        #print("Done with this category")
    else: 
        for position_children in position['children']:
            for job in position_children['jobs']:
                database.append(greenhouse_fetch_data(job))
            #print("Done with this CHILDREN category")
            
        for job in position['jobs']: 
            database.append(greenhouse_fetch_data(job))
        #print("Done with this category")
```
