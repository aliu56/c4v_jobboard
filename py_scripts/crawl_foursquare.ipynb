# get_country
get_country returns the country of a city or the country itself if it is a country 
Whatever the input language, answer is returned in english

```python
from geopy.geocoders import Nominatim
 
geolocator = Nominatim(user_agent = "geoapiExercises")

def get_country(place):
    res = str(geolocator.geocode(place, language = "en"))
    country = res.split(',')[-1]
    return country
```

```python
soup = BeautifulSoup(source, 'html.parser')

print(soup.prettify())
```

```python
database = [['job_name', 'country', 'city', 'offer_link']]

positions = soup.find_all('div', class_ = 'opening')

for position in positions:
    print("______________")
    print(position)
    
    #JOB_NAME
    job_name = position.find("a").text.strip(' \n\t')
    #print(job_name)        
    
    # COUNTRY / CITY
    if position.find("span").text == '':
        city = 'Remote'
        country = 'Remote'
    else: 
        city = position.find("span").text.strip(' \n\t')
        country = get_country(city).strip(' \n\t')
        
#     print(country)
#     print('the city is: '+ city)
    
    # OFFER_URL
    url = position.a.get('href')    
    offer_url = "https://boards.greenhouse.io" + url
    print(offer_url)


    database.append([job_name, country, city, offer_url])
```
